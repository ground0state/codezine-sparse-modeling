```python
import pandas as pd
```

```python
data = pd.read_csv('sales.csv')
data
```

```python
def print_model(model, X, y):
    print(f"Coefficients: {list(model.coef_)}")
    print(f"Intercept: {model.intercept_}")
    print(f"R square: {model.score(X, y)}")
```

```python
y = data["売上 (千円)"]
X = data.iloc[:, 1:-1]
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
print_model(lr, X, y)
```

```python
from sklearn.linear_model import Lasso

clf = Lasso(alpha=5)
clf.fit(X, y)
print_model(clf, X, y)
```

```python
from sklearn.linear_model import OrthogonalMatchingPursuit

omp = OrthogonalMatchingPursuit(n_nonzero_coefs=4)
omp.fit(X, y)
print_model(omp, X, y)
```
